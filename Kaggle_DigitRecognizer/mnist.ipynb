```python
def norm_pixels(x):
    x = x/255.0
    return x
```

```python
xtrainNorm = norm_pixels(train.iloc[:,1:].to_numpy())
```

```python
xtestNorm = norm_pixels(test.to_numpy())
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(drop = None)
ytrain = ohe.fit_transform(train[['label']]).toarray()


xtrain = xtrainNorm

xtrain.shape,ytrain.shape

#xval.shape, yval.shape

ytrain
```

**NN**

```python
from tensorflow import keras
from tensorflow.keras import layers
```

```python
NNmodel = keras.Sequential(
[
    layers.Dense(units= 512,activation = 'relu',input_shape=xtrain.shape[1:]),
    layers.Dense(units= 128,activation = 'relu'),
    layers.Dense(units= 32,activation = 'relu'),
    layers.Dense(units= 10,activation = 'softmax')
]
)


NNmodel.summary()
```

```python
NNmodel.compile(
    optimizer=keras.optimizers.Adam(),  
    loss=keras.losses.CategoricalCrossentropy(),
    metrics=[keras.metrics.CategoricalAccuracy()],
)
```

```python
history = NNmodel.fit(
    xtrain,
    ytrain,
    batch_size=512,
    epochs=15,
    shuffle = True,
    validation_split = 0.3
)
```

```python
# testing
ypred_ohe = NNmodel.predict(xtestNorm)
ypred_ohe.shape
```

```python
ypred = np.argmax(ypred_ohe, axis=1)
ypred
```

```python
y_out = pd.DataFrame(ypred, index = range(1,ypred.shape[0]+1),columns = ['Label'])
y_out.index.name = 'ImageId'
y_out.reset_index(inplace=True)  
```
